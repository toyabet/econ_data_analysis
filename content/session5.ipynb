### 絶対値関数

```python
abs(-1)
```

```python
def user_abs(x):
    if x < 0:
        x = -x
    return x
```

```python
user_abs(-1)
```

### len関数

```python
snacks = ["バナナ", "りんご", "ラムネ"]
print(len(snacks))
```

```python
stock = {
    "apple": 10,
    "banana": 0,
    "orange": 5
}
print(len(stock))
```

```python
# lenを文字列に使うと、文字数を取得する
message = "Python"
print(len(message))
```

### 引数を複数持つ関数

```python
def rectangle_area(width, height):
    return width * height

print(rectangle_area(3, 4))
print(rectangle_area(10, 2.5))
```

### デフォルト引数

```python
# 引数には、あらかじめ標準の値を設定できる

def add_tax(price, tax_rate=0.10):
    return price * (1 + tax_rate)

print(add_tax(1000))
print(add_tax(1000, 0.08))
```

### キーワード引数

```python
print(add_tax(1000, tax_rate=0.08))
```

## モジュールのインポート

```python
import math

math.sqrt(16)
```

#### `from ... import ...` で関数を直接実行する

```python
from math import sqrt

print(sqrt(16))
```

#### `import ... as ...` でモジュール名に別名をつける

```python
import math as m

print(m.sqrt(25))
```
