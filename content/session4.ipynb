### 入れ子構造のif文

```python
stock = {
    "apple": 10,
    "banana": 0,
    "orange": 5
}

item = "apple"
quantity = 3

if item in stock:
    if stock[item] >= quantity:
        stock[item] = stock[item] - quantity
        # 本当なら、この部分に売上計上や発送処理などが入る
        print("注文を受け付けました")
    else:
        print("在庫が不足しています")
else:
    print("その商品は取り扱っていません")
stock
```

## for文による繰り返し

```python
numbers = [1, 2, 3, 4, 5]

# iを1から5まで、5回繰り返して出力する
for i in numbers:
    print(i)
```

# while文による繰り返し

```python
#　x　の平方根を求める
x = 2
rnew = x
diff = rnew - x/rnew
# r1 と r2 の差が 0.000001 より大きい間は繰り返す
while (abs(diff) > 1.0E-6):
    r1 = rnew
    r2 = x/r1
    rnew = (r1 + r2)/2
    print(r1, rnew, r2)
    diff = r1 - r2
rnew
```
