# 経済成長の分析 ― 実質GDPの長期時系列をよむ

経済成長とは、一国の経済の規模が時間とともに大きくなることをいう。その規模を測る代表的な指標が国内総生産（GDP）である。

都道府県別データのように、ある一時点で多数の対象を並べたデータを横断面データという。これに対して、GDPの推移のように、1つの対象を時間にそって並べたデータを時系列データという。

内閣府「国民経済計算」のGDPを用いて、戦後日本の経済成長を時系列データとして分析する。

## 1. 今回の問い

日本の実質GDPは、戦後どのように変化してきたか。そして、その成長の勢いは時代によってどう異なってきたか。

## 3. 使用するデータ

内閣府「国民経済計算」より、日本の名目GDPと実質GDP成長率を用いる。期間は1955年から2020年の暦年である。

GDPには名目と実質の2つがある。名目GDPは、その年の価格でそのまま評価した金額である。物価が上がると、同じだけ生産しても名目GDPは増える。実質GDPは、ある基準の年の価格を使って毎年の生産を評価しなおしたGDPであり、物価全体の上昇による見かけの増加を取り除いたものである。そのため、実質GDPの変化は、物価上昇による分を除いた生産の動きを表す。経済成長は、この物価の影響を除いた生産の動きを見るため、実質でとらえる。

名目GDPは総額（10億円）、実質GDP成長率は実質GDPの前年比（％）である。前年比成長率は前の年と比べて計算するため、データの最初の年である1955年には前年がなく、計算できない。このため1955年の成長率は欠損（NaN）とする。前年比をとると、成長率が得られる年数は元のGDPの年数より1つ少なくなる。

長期系列は、複数の基準年の計数を接続したものである。

```python
# 日本の名目GDPと実質GDP成長率（暦年、1955〜2020年）
# 出典：内閣府「国民経済計算」長期経済統計（令和3年度 年次経済財政報告）
# URL：https://www5.cao.go.jp/j-j/wp/wp-je21/h11_data01.html
# 名目GDP：総額（10億円）
# 実質成長率：実質GDPの前年比（％）。1955年は基準年のため NaN

year = list(range(1955, 2021))

nominal_gdp = [
    8897.3, 10016.4, 11543.1, 12266.0, 14022.2, 17019.4, 20556.0, 23326.5, 26697.0, 31404.3,
    34938.7, 40577.2, 47551.5, 56315.8, 66153.4, 77970.5, 85790.8, 98221.3, 119592.9, 142710.0,
    157681.5, 177078.4, 197328.4, 217295.0, 235518.6, 255322.8, 273857.6, 287866.1, 300825.9, 319000.9,
    339925.7, 356896.0, 373189.4, 401002.2, 429350.6, 462090.7, 491874.3, 504313.0, 504552.6, 510916.1,
    521613.5, 535562.1, 543545.4, 536497.4, 528069.9, 535417.7, 531653.9, 524478.7, 523968.6, 529400.9,
    532515.6, 535170.2, 539281.7, 527823.8, 494938.4, 505530.6, 497448.9, 500474.7, 508700.6, 518811.0,
    538032.3, 544364.6, 553073.0, 556189.6, 559862.3, 538697.4
]

real_growth = [
    np.nan, 7.5, 7.8, 6.2, 9.4, 13.1, 11.9, 8.6, 8.8, 11.2,
    5.7, 10.2, 11.1, 11.9, 12.0, 10.3, 4.4, 8.4, 8.0, -1.2,
    3.1, 4.0, 4.4, 5.3, 5.5, 2.8, 4.2, 3.3, 3.5, 4.5,
    5.2, 3.3, 4.7, 6.8, 4.9, 4.9, 3.4, 0.8, -0.5, 1.0,
    2.6, 3.1, 1.0, -1.3, -0.3, 2.8, 0.4, 0.0, 1.5, 2.2,
    1.8, 1.4, 1.5, -1.2, -5.7, 4.1, 0.0, 1.4, 2.0, 0.3,
    1.6, 0.8, 1.7, 0.6, 0.0, -4.6
]
```

```python
df = pd.DataFrame({
    "year": year,
    "nominal_gdp": nominal_gdp,
    "real_growth": real_growth,
})

# 名目GDPを兆円に直す
df["nominal_gdp_tril"] = df["nominal_gdp"] / 1000

df.head()
```

```python
# 行数と列数を確認する
print(df.shape)
```

## 4. 横断面データと時系列データ

データは、その並び方によって横断面データと時系列データに分けられる。それぞれの形を、実際の表で見比べる。

横断面データは、ある一時点で多数の対象を横に並べたデータである。これまで用いてきた都道府県データがこれにあたる。次の表は、ある年の都道府県別の最低賃金と物価水準であり、1行が1つの都道府県を表す。

```python
# 例：横断面データ（都道府県別、ある年）
example_cross = pd.DataFrame({
    "prefecture": ["北海道", "東京", "愛知", "大阪", "沖縄"],
    "minimum_wage": [1075, 1226, 1140, 1177, 1023],
    "price_index": [101.9, 104.0, 98.1, 99.3, 100.2],
})
example_cross
```

時系列データは、1つの対象を時間にそって繰り返し観測して並べたデータである。今回のGDPがこれにあたる。次の表は日本のGDPであり、1行が1つの年を表す。

```python
# 時系列データ（日本、1955〜2020年）
df[["year", "nominal_gdp_tril", "real_growth"]].head()
```

2つの表は、1行が表すものが異なる。横断面データでは1行が1つの対象（都道府県）、時系列データでは1行が1つの時点（年）である。

| 比べる点 | 横断面データ | 時系列データ |
|---|---|---|
| 例 | 都道府県別の最低賃金 | 日本の実質GDP |
| 1行が表すもの | 1つの対象（都道府県） | 1つの時点（年） |
| 横に並ぶもの | 多数の対象 | 多数の時点 |
| 観測の時点 | ある一時点 | 1955〜2020年 |
| 主な関心 | 対象どうしの違い | 時間に伴う変化 |
| 行の並べ替え | 入れ替えてよい | 入れ替えられない |

横断面データでは対象の並び順に意味がなく、入れ替えても分析は変わらない。一方、時系列データでは年に順序があり、入れ替えることはできない。前年比成長率のように隣り合う年を比べる見方や、長期的な傾向を読み取る見方は、順序のある時系列データだからこそ成り立つ。グラフにするときは、横軸に年をとる。

## 5. 名目GDPの推移を折れ線で見る

```python
plt.figure(figsize=(9, 5))
plt.plot(df["year"], df["nominal_gdp_tril"])
plt.title("日本の名目GDPの推移")
plt.xlabel("年")
plt.ylabel("名目GDP（兆円）")
plt.grid(True)
plt.show()
```

名目GDPは戦後から大きく増えている。ただし、この増加には物価の上昇も含まれる。

## 6. 前年比成長率を計算する

成長率は、前の年とくらべて何％増えたかを表す。名目GDPの前年比成長率は、次の式で計算する。

$$
g_t = \frac{Y_t - Y_{t-1}}{Y_{t-1}} \times 100
$$

pandas では `pct_change()` で前年比を計算できる。

```python
# 名目GDPの前年比成長率（％）
df["nominal_growth"] = df["nominal_gdp"].pct_change() * 100

df[["year", "nominal_gdp_tril", "nominal_growth"]].head()
```

```python
plt.figure(figsize=(9, 5))
plt.bar(df["year"], df["nominal_growth"])
plt.title("名目GDPの前年比成長率")
plt.xlabel("年")
plt.ylabel("成長率（％）")
plt.axhline(0, color="black", linewidth=0.8)
plt.grid(True, axis="y")
plt.show()
```

名目では20％近く伸びた年もあるが、これには物価の上昇が含まれる。

## 7. 名目と実質：1974年の衝撃

名目GDPの前年比（名目成長率）と実質GDPの前年比（実質成長率）を並べると、両者が大きく食い違う年がある。

```python
plt.figure(figsize=(9, 5))
plt.plot(df["year"], df["nominal_growth"], marker="o", markersize=3, label="名目成長率")
plt.plot(df["year"], df["real_growth"], marker="o", markersize=3, label="実質成長率")
plt.title("名目成長率と実質成長率")
plt.xlabel("年")
plt.ylabel("成長率（％）")
plt.axhline(0, color="black", linewidth=0.8)
plt.legend()
plt.grid(True)
plt.show()
```

1974年は第一次石油危機（オイルショック）で物価が大きく上がった年である。

```python
df[df["year"] == 1974][["year", "nominal_growth", "real_growth"]]
```

名目成長率は約+19％だが、実質成長率は約−1％である。物価は大きく上がったが、ものやサービスの生産はむしろ減っていた。実質を用いるのは、こうした物価の影響を取り除くためである。

## 8. 実質GDPの水準をつくる

実質成長率は前年比で何％増えたかを表す。1年で g％ 成長すると、その年のGDPは前年の (1 + g/100) 倍になる。この倍率を毎年かけ合わせると、実質GDPの水準の動きが得られる。ここでは1955年を100とした指数で表す。

```python
# 実質成長率から、実質GDPの水準（1955年=100の指数）を作る
factor = 1 + df["real_growth"].fillna(0) / 100
df["real_index"] = 100 * factor.cumprod()

df[["year", "real_growth", "real_index"]].head()
```

```python
plt.figure(figsize=(9, 5))
plt.plot(df["year"], df["real_index"])
plt.title("実質GDPの水準（1955年=100）")
plt.xlabel("年")
plt.ylabel("実質GDP（1955年=100）")
plt.grid(True)
plt.show()
```

実質GDPは長期に増え続け、1973年には1955年の約5倍に達している。ただし近年はカーブがゆるやかで、水準のグラフだけでは成長の勢いの差がとらえにくい。

## 9. 対数スケールで見る

これまでのグラフは、縦軸が普通の目盛（線形目盛）であった。線形目盛では、同じ間隔が同じ差を表す。たとえば100から200までと、300から400までは、どちらも差が100なので、同じ間隔になる。

対数スケールでは、同じ間隔が同じ倍率（割合）を表す。たとえば100から200まで（2倍）と、200から400まで（2倍）が同じ間隔になる。目盛は、次のように10倍ごとに等間隔で並ぶ。

| 普通の目盛（線形） | 対数スケール |
|---|---|
| 100, 200, 300, …（同じ差ずつ） | 100, 1000, 10000, …（同じ倍率ずつ） |

経済成長で問題になるのは、何円増えたかよりも、前年から何％（何倍）増えたかである。対数スケールは、この倍率の変化を見るのに向いている。

一定の割合で増え続ける量は、普通の目盛では右上にいくほど急に立ち上がる曲線になる。同じものを対数スケールで描くと直線になり、その傾きが成長率に対応する。傾きが急なほど成長が速い。

```python
plt.figure(figsize=(9, 5))
plt.plot(df["year"], df["real_index"])
plt.yscale("log")   # 縦軸を対数スケールにする
plt.title("実質GDPの水準（対数スケール、1955年=100）")
plt.xlabel("年")
plt.ylabel("実質GDP（対数目盛）")
plt.grid(True, which="both")
plt.show()
```

8節の普通の目盛のグラフでは、近年の大きな値が目立ち、初期の動きはほとんど平らに見えた。対数スケールにすると、各時代の伸びの勢いを同じ図の中で比べられる。1955年から1970年代はじめまでは傾きが急で、成長が速かった。1990年代以降は傾きがゆるやかで、成長は遅い。

## 10. 時代を区切って平均成長率をくらべる

期間を3つに区切り、それぞれの実質成長率の平均をとる。

- 高度経済成長期：1956〜1973年
- 安定成長期：1974〜1990年
- 1991年以降：1991〜2020年（いわゆる「失われた」時期）

```python
# ある期間の実質成長率の平均を求める関数
def average_growth(start, end):
    period = df[(df["year"] >= start) & (df["year"] <= end)]
    return period["real_growth"].mean()

# 3つの時代について平均を計算する
mean_high = average_growth(1956, 1973)    # 高度成長期
mean_stable = average_growth(1974, 1990)  # 安定成長期
mean_recent = average_growth(1991, 2020)  # 1991年以降

print("高度成長期 (1956-1973):", round(mean_high, 1), "%")
print("安定成長期 (1974-1990):", round(mean_stable, 1), "%")
print("1991年以降 (1991-2020):", round(mean_recent, 1), "%")
```

```python
names = ["高度成長期\n1956-1973", "安定成長期\n1974-1990", "1991年以降\n1991-2020"]
means = [mean_high, mean_stable, mean_recent]

plt.figure(figsize=(8, 5))
plt.bar(names, means, color=["tab:red", "tab:orange", "tab:blue"])
plt.title("時代別の平均実質成長率")
plt.ylabel("平均成長率（％）")

# 各棒の上に数値を表示する
plt.text(0, mean_high + 0.2, f"{mean_high:.1f}%", ha="center")
plt.text(1, mean_stable + 0.2, f"{mean_stable:.1f}%", ha="center")
plt.text(2, mean_recent + 0.2, f"{mean_recent:.1f}%", ha="center")

plt.grid(True, axis="y")
plt.show()
```

平均成長率は、高度成長期の約9％から、安定成長期の約4％、1991年以降の1％程度へと、時代とともに大きく低下している。

## 11. 移動平均でならす（発展）

単年の成長率は景気によって上下に大きく動く。何年かの平均をとって動きをなめらかにしたものを移動平均という。ここでは5年の移動平均をとる。

```python
# 5年移動平均
df["ma5"] = df["real_growth"].rolling(5).mean()

plt.figure(figsize=(9, 5))
plt.plot(df["year"], df["real_growth"], alpha=0.4, label="実質成長率（各年）")
plt.plot(df["year"], df["ma5"], label="5年移動平均")
plt.title("実質成長率と5年移動平均")
plt.xlabel("年")
plt.ylabel("成長率（％）")
plt.axhline(0, color="black", linewidth=0.8)
plt.legend()
plt.grid(True)
plt.show()
```

5年移動平均でみると、成長率は長期にわたって段階的に低下している。

## 12. まとめ

日本の実質GDPは戦後一貫して増加した。一方で成長の勢いは時代とともに低下し、実質成長率の平均は高度成長期の約9％から、安定成長期の約4％、1991年以降の1％程度へと下がった。

成長は、物価変動を除いた実質で測る。名目と実質は、1974年のように大きく異なることがある。

ここでの分析は成長のすがたを記述したものであり、成長率が低下した原因を示すものではない。要因を変数の関係として調べるには回帰分析を用いる。対数をとった実質GDPを時間に回帰すると、その傾きが平均成長率に対応する。
