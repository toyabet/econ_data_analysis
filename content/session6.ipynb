## 折れ線グラフ

### 使用するデータ

```python
# 消費者物価指数、総合、年平均
# 出典：総務省統計局「2020年基準 消費者物価指数、全国、2025年平均」
# URL：https://www.stat.go.jp/data/cpi/sokuhou/nen/pdf/zen-n.pdf
# 注：指数の基準は 2020年＝100

years = [2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023, 2024, 2025]
overall_cpi = [94.9, 97.5, 98.2, 98.1, 98.6, 99.5, 100.0, 100.0, 99.8, 102.3, 105.6, 108.5, 111.9]
```

```python
# リストの中身を確認
print(years)
print(overall_cpi)
```

```python
print(years[0])
print(overall_cpi[0])
```

### 折れ線グラフをプロットする

```python
plt.plot(years, overall_cpi)
plt.show()
```

```python
plt.plot(years, overall_cpi, marker="o")
plt.title("消費者物価指数の推移 (総合)")
plt.xlabel("年")
plt.ylabel("総合指数、2020年＝100")
plt.xticks(years, rotation=45)
plt.grid(axis="y", linestyle="--", alpha=0.6)
plt.show()
```

## 棒グラフ

### 使用するデータ1

```python
# 世帯主の年齢階級別消費支出、二人以上の世帯、2025年
# 出典：総務省統計局「家計調査報告 家計収支編 2025年平均結果の概要」図Ⅰ－１－４、表Ⅰ－１－２
# URL：https://www.stat.go.jp/data/kakei/sokuhou/tsuki/pdf/fies_gaikyo2025.pdf
# 単位：1世帯当たり1か月平均、円

age_groups = ["40歳未満", "40〜49歳", "50〜59歳", "60〜69歳", "70歳以上"]
consumption = [299100, 348607, 367643, 327405, 264332]
```

### 棒グラフをプロットする1

```python
plt.bar(age_groups, consumption)

plt.title("世帯主の年齢階級別消費支出 (2025年)")
plt.xlabel("世帯主の年齢階級")
plt.ylabel("月平均額、円")

plt.grid(axis="y", linestyle="--", alpha=0.6)

plt.show()
```

```python
plt.bar(age_groups, consumption, width=0.6)

plt.title("世帯主の年齢階級別消費支出 (2025年)")
plt.xlabel("世帯主の年齢階級")
plt.ylabel("月平均額、円")

plt.grid(axis="y", linestyle="--", alpha=0.6)
plt.ylim(0, 400000)

plt.show()
```

### 使用するデータ2

```python
# 出典： 総務省統計局「家計調査報告〔家計収支編〕2025年平均」
# 表3「消費支出の内訳、2025年平均、二人以上の世帯」
# URL：https://www.stat.go.jp/data/kakei/sokuhou/tsuki/pdf/fies_mr-y.pdf

categories = [
    "食料",
    "住居",
    "光熱・水道",
    "家具・家事用品",
    "被服及び履物",
    "保健医療",
    "交通・通信",
    "教育",
    "教養娯楽",
    "その他の消費支出",
]

expenditure = [
    94895,
    18678,
    24547,
    13068,
    10063,
    15863,
    45730,
    11939,
    32125,
    47093,
]
```

### 棒グラフをプロットする2

```python
plt.figure(figsize=(8, 6))

plt.barh(categories, expenditure, height=0.6)

plt.title("消費支出の内訳、二人以上の世帯 (2025年)")
plt.xlabel("月平均額、円")
plt.ylabel("費目")

plt.gca().invert_yaxis()
plt.grid(axis="x", linestyle="--", alpha=0.4)

plt.show()
```

## ヒストグラム

### 使用するデータ

```python
# 都道府県別データ
# 出典1：厚生労働省「令和7年度 地域別最低賃金 全国一覧」
# URL：https://www.mhlw.go.jp/content/11200000/001571192.pdf
#
# 出典2：総務省統計局「消費者物価地域差指数－小売物価統計調査（構造編）2024年（令和6年）結果－」別表1
# URL：https://www.stat.go.jp/data/kouri/kouzou/pdf/g_2024.pdf
# 消費者物価地域差指数は、地域間の物価水準の差を表す指数

prefectures = [
    "北海道", "青森", "岩手", "宮城", "秋田", "山形", "福島",
    "茨城", "栃木", "群馬", "埼玉", "千葉", "東京", "神奈川",
    "新潟", "富山", "石川", "福井", "山梨", "長野", "岐阜",
    "静岡", "愛知", "三重", "滋賀", "京都", "大阪", "兵庫",
    "奈良", "和歌山", "鳥取", "島根", "岡山", "広島", "山口",
    "徳島", "香川", "愛媛", "高知", "福岡", "佐賀", "長崎",
    "熊本", "大分", "宮崎", "鹿児島", "沖縄"
]

minimum_wages = [
    1075, 1029, 1031, 1038, 1031, 1032, 1033,
    1074, 1068, 1063, 1141, 1140, 1226, 1225,
    1050, 1062, 1054, 1053, 1052, 1061, 1065,
    1097, 1140, 1087, 1080, 1122, 1177, 1116,
    1051, 1045, 1030, 1033, 1047, 1085, 1043,
    1046, 1036, 1033, 1023, 1057, 1030, 1031,
    1034, 1035, 1023, 1026, 1023
]

price_level_index = [
    101.9, 98.5, 100.0, 100.6, 99.2, 101.4, 98.8,
    97.5, 97.6, 96.2, 100.3, 101.2, 104.0, 103.3,
    98.0, 98.6, 99.5, 99.3, 97.7, 97.9, 97.1,
    98.3, 98.1, 98.7, 98.6, 101.1, 99.3, 99.2,
    98.1, 98.2, 98.9, 100.0, 97.7, 98.7, 99.9,
    99.3, 98.6, 98.6, 100.0, 98.0, 97.7, 99.3,
    99.4, 97.4, 97.0, 96.4, 100.2
]
```

### ヒストグラムをプロットする

```python
# 50円刻みで区間を作る
bins = [1000, 1050, 1100, 1150, 1200, 1250]

plt.hist(minimum_wages, bins=bins, rwidth=0.98)

plt.title("都道府県別最低賃金の分布 (令和7年度)")
plt.xlabel("最低賃金、円")
plt.ylabel("都道府県数")
plt.xticks(bins)

plt.grid(axis="y", linestyle="--", alpha=0.6)

plt.show()
```

## 散布図

```python
plt.scatter(price_level_index, minimum_wages, alpha=0.75)

plt.title("物価水準と最低賃金 (都道府県別)")
plt.xlabel("消費者物価地域差指数、全国平均＝100 (2024年)")
plt.ylabel("最低賃金 (令和7年度)")

plt.axvline(100, linestyle="--", alpha=0.5)
plt.grid(True, linestyle="--", alpha=0.4)
plt.xlim(95.5, 104.5)
plt.ylim(1000, 1250)

plt.show()
```

```python
plt.figure(figsize=(9, 6))

plt.scatter(price_level_index, minimum_wages, alpha=0.75)

label_prefectures = ["東京", "神奈川", "大阪", "北海道", "群馬", "鹿児島", "沖縄"]

for i in range(len(prefectures)):
    if prefectures[i] in label_prefectures:
        plt.text(price_level_index[i] + 0.05, minimum_wages[i] + 2, prefectures[i], fontsize=9)

plt.title("物価水準と最低賃金、都道府県別")
plt.xlabel("消費者物価地域差指数、全国平均＝100 (2024年)")
plt.ylabel("最低賃金 (令和7年度)")

plt.axvline(100, linestyle="--", alpha=0.5)
plt.grid(True, linestyle="--", alpha=0.4)
plt.xlim(95.5, 104.5)
plt.ylim(1000, 1250)

plt.show()
```
